### Conclusion
There are obvious noise in the beginning and end of the recording of the accelerometer data. This is due to the movement of starting/ending the recording, putting/picking the phone in/from the backpocket.

## Removal of noise in the beginning and end

```python
x = data_picking_raw[["ACCELEROMETER X (m/s²)"]]
x = x - x.mean()
x = x[50:-100]

plt.figure(figsize=(20,5))
plt.plot(x)
```

```python
x = data_picking_raw[["ACCELEROMETER Y (m/s²)"]]
x = x - x.mean()
x = x[50:-100]

plt.figure(figsize=(20,5))
plt.plot(x)
```

```python
x = data_picking_raw[["ACCELEROMETER Z (m/s²)"]]
x = x - x.mean()
x = x[50:-100]

plt.figure(figsize=(20,5))
plt.plot(x)
```

### Conclusion:
The direction of the movement depends on the position of the phone. Looking at the plots we see high negative accelerations in the Y-direction. This indicates the phone have been held in an up-right position when picking things up, which is reasonable due to the placement in the backpocket. 

## Labelling of picking sequence
It is difficult to say how long a picking sequence is, but looking at the mean acceleration in y-direction over a time-window of 1 second (10 datapoints) will increase the likelihood of detecting and separating the picking and walking movements.

```python
x = data_picking_raw[["ACCELEROMETER Y (m/s²)"]]
x = x - x.mean()
x = x[50:-100] 

x.head()
```

```python
x_grouped = x.groupby(np.arange(len(x))//10).mean()
x_grouped = x_grouped - x_grouped.mean()

plt.plot(range(len(x_grouped)), x_grouped)
```

### Conclusion:
The time window can be further investigated, but for now we believe having a threshold above 0.5 and below -0.5 as an average acceleration in y-direction will help us distinguish between picking and walking.

## Final output

```python
def set_activity(accelerometer_y):
    activity = 'walking'
    if np.abs(accelerometer_y) > 0.5:
        activity = 'picking'
    return activity
```

```python
x_grouped['activity'] = x_grouped['ACCELEROMETER Y (m/s²)'].apply(set_activity)
```

```python
x_grouped['activity'].value_counts()
```

We have 185 data points in the picking dataset, whereof 58 seconds (sequences) will be labelled as picking and 127 seconds will be labelled as walking. One can argue if 58 picking sequences is reasonable in the 4 minute timespan. In reality maybe not, but for labelling purposes we wanted to collect as much picking sequences as possible to work with.
